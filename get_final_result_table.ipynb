```python
import pickle 
import pandas as pd
target_path = './past_results_pictures/result_September_17_new2.pkl'
def format_ablation_results(df):
    """
    Processes a raw evaluation DataFrame to match the ablation study table format from the paper.
    Groups results by 'K', calculates the mean, renames columns, and formats the output.
    """
    # 1. Isolate relevant columns: keep only 'nDCG' metric columns and the 'K' index
    cols = [col for col in df.columns if 'nDCG' in col] + ["K"]
    
    # 2. Aggregate the data: group by 'K' and calculate the mean for the selected columns
    grouped_result = df[cols].groupby('K').agg('mean')
    
    # 3. Define the column mapping to match the naming convention in the paper
    rename_mapping = {
        'baseline_nDCG': 'Spatial Only',
        'no_masking_nDCG': 'No Personalization',
        'no_temporal_nDCG': 'No Temporal',
        'full_nDCG': 'Full Model'
    }
    
    # 4. Rename the columns using the mapping
    grouped_result = grouped_result.rename(columns=rename_mapping)
    
    # 5. Filter and reorder the columns to match the paper's exact left-to-right layout
    paper_columns = ['Spatial Only', 'No Personalization', 'No Temporal', 'Full Model']
    final_result = grouped_result[paper_columns]
    
    # 6. Round the values to 4 decimal places to perfectly mimic the paper's formatting
    return final_result.round(4)


# ==========================================
# Main Execution
# ==========================================

# Load the raw results from the pickle file
with open(target_path, 'rb') as f:
    data = pickle.load(f)

# Convert the raw dictionaries into pandas DataFrames
nyc_df = pd.DataFrame(data['NYC'])
tky_df = pd.DataFrame(data['TKY'])

# Apply our formatting function to both datasets
nyc_result = format_ablation_results(nyc_df)
tky_result = format_ablation_results(tky_df)

# Display the final tables
print("=== NYC Dataset ===")
print(nyc_result)
print("\n=== Tokyo Dataset ===")
print(tky_result)
```
